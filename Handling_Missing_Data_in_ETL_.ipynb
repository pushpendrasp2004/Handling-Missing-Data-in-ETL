1. What are the most common reasons for missing data in ETL pipelines?

- Missing data in ETL pipelines occurs due to several reasons:

Data Entry Errors
Users may skip fields or enter incomplete information during data collection.

System Integration Issues
When data is extracted from multiple sources, some fields may not exist in certain systems.

Data Transmission Failures
Network interruptions or corrupted files may cause loss of data during transfer.

Schema Mismatch
Differences between source and target database structures can result in missing fields.

Optional Fields in Forms
Some fields are not mandatory, so users may leave them blank.

Sensor or Logging Failures
In automated systems, hardware or software failures may prevent data capture.

Missing data is a common challenge in ETL workflows and must be handled carefully to ensure data quality.

2. 2. Why is blindly deleting rows with missing values considered bad practice in ETL?

Blindly deleting rows with missing values is considered poor practice because:

Loss of Valuable Information
Important data might be removed unnecessarily.

Reduced Dataset Size
Too many deletions can reduce the dataset and affect analysis.

Bias in Results
If missing values occur in specific groups, deleting them may distort the results.

Incorrect Business Decisions
Removing data without analysis may lead to misleading insights.

Therefore, analysts should evaluate the the reason for missing values before removing data.

3. Difference Between Listwise Deletion and Column Deletion
- | Method                | Description                                         | When Appropriate                                                        |
| --------------------- | --------------------------------------------------- | ----------------------------------------------------------------------- |
| **Listwise Deletion** | Entire rows are removed if any value is missing     | When missing data is small and random                                   |
| **Column Deletion**   | Entire column is removed if many values are missing | When most values in a column are missing and the column is not critical |

Example Scenarios

Listwise Deletion Scenario:
If only 2–3 rows in a dataset of 10,000 rows have missing values.

Column Deletion Scenario:
If a column has more than 80–90% missing values and is not important for analysis.

4. Why is Median Imputation Preferred Over Mean for Skewed Data?

- Median imputation is preferred because:

Median is not affected by extreme values (outliers).

Mean can shift significantly if very high or very low values exist.

In skewed datasets like income, a few extremely rich individuals can distort the mean.

Example:

Income values
20000, 22000, 24000, 25000, 200000

Mean = very high due to outlier
Median = realistic central value

Thus median represents the typical value better.

5. What is Forward Fill and Where is it Useful?

Forward Fill (FFILL) is a technique where missing values are replaced using the previous valid value in the dataset.

| Date | Sales |
| ---- | ----- |
| Day1 | 100   |
| Day2 | NaN   |
| Day3 | NaN   |
| Day4 | 150   |

After Forward Fill:

| Date | Sales |
| ---- | ----- |
| Day1 | 100   |
| Day2 | 100   |
| Day3 | 100   |
| Day4 | 150   |

Most useful in:

Time-series datasets

Stock prices

Daily sales data

Sensor readings

6. Why Should Missing Values be Flagged Before Imputation?

Flagging missing values before imputation is important because:

Preserves Information
Analysts can still know which values were originally missing.

Improves Model Accuracy
Machine learning models can use the missing flag as a feature.

Business Insights
Missing values may reveal patterns (e.g., customers not reporting income).

Example flag column:

| Income | Income_Missing_Flag |
| ------ | ------------------- |
| 50000  | 0                   |
| NaN    | 1                   |

7. How Can Missing Income Provide Business Insights?

Missing income values can indicate:

Customers hesitant to share financial data

Certain demographic groups avoiding disclosure

High-income customers protecting privacy

Incomplete data collection processes

Businesses can use this insight to:

Improve form design

Target marketing strategies

Segment customers differently

Thus missing data itself can reveal behavioral patterns.
